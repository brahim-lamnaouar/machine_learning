```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
import numpy as np 
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot as plt
```

```
data = load_iris()
print(data.feature_names)
X = data.data
y = data.target
print(data.target_names)
```

```
print(X.shape)
```

```
v1=0
v2=2
plt.scatter(X[:,v1],X[:,v2],c=y)
plt.xlabel(data.feature_names[v1])
plt.ylabel(data.feature_names[v2])

```

```
knn=KNeighborsClassifier(n_neighbors=7)
scores = cross_val_score(knn, X, y ,cv=5, scoring="accuracy")
print(scores)
print(scores.mean())
```

```
k_range=range(1,31)
k_scores=[]

for k in  k_range:
  knn=KNeighborsClassifier(n_neighbors=k, metric="minkowski")
  scores = cross_val_score(knn, X, y ,cv=5, scoring="accuracy")
  k_scores.append(scores.mean())
  print(k,"  ", scores.mean())


plt.plot(k_range,k_scores)
```

```
from sklearn.model_selection import GridSearchCV
k_range = list(range(1,31))
metrics=["manhatan","euclidean","minkowski"]

knn = KNeighborsClassifier(n_neighbors=1)
param_grid = dict(n_neighbors= k_range, metric= metrics)
grid=GridSearchCV(knn, param_grid ,cv= 5, scoring = "accuracy")

grid.fit(X,y)
```

```
print(grid.best_estimator_)
```

```
print(grid.best_estimator_.score(X,y))
```
